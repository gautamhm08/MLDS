```python
Python String is a sequence of Unicode characters used to store and manipulate text data.  It is one of the most commonly used data types in Python for handling names, messages, file content, user input, and more. 

Strings are created by enclosing text in single quotes ('...'), double quotes ("..."), or triple quotes ('''...''' or """...""") for multi-line strings. 
triple quotes ('''...''' or """...""") for multi-line strings is also called as ocstrings (documentation strings).
Example: greeting = "Hello, World!" or name = 'Alice'.

Strings are immutable, meaning once created, their content cannot be changed.  Any modification results in a new string.

Unicode support allows strings to handle characters from any language, symbols, and emojis.


Common operations include:
1.Indexing and slicing to access individual characters or substrings (e.g., text[0] or text[1:4]).

2.Concatenation using + (e.g., "Hello" + " " + "World").

3.Repetition using * (e.g., "Hi" * 3 → "HiHiHi").

4.Membership testing with in (e.g., "Py" in "Python" → True). 

5.Built-in methods simplify text processing:
.upper(), .lower() – change case.
.strip() – remove whitespace.
.replace(old, new) – replace text.
.split() – split into a list.
.join() – combine list elements into a string.

String formatting is done using:
f-strings (preferred): f"Hello, {name}!".
```

```python
text = "python Programming is FUN"

print(text.upper())           # PYTHON PROGRAMMING IS FUN
print(text.lower())           # python programming is fun
print(text.capitalize())      # Python programming is fun
print(text.title())           # Python Programming Is Fun
print(text.swapcase())        # PYTHON pROGRAMMING IS fun # exchange lowercase and upper case

# Case checking
print(text.isupper())         # False
print(text.islower())         # False
print("123".islower())        # False (numbers don't count)
```

```python
Search and Replace Methods
```

```python
text = "Python is amazisng. Python is powerful."

# Finding substrings
print(text.find("Python"))      # 0 (first occurrence)
print(text.find("Java"))        # -1 (not found)
print(text.rfind("Python"))     # 19 (last occurrence)
print(text.index("Python"))     # 0 (similar to find but raises error if not found)
print(text.count("is"))         # 2
print(text.count("python"))
# Checking starts/ends
print(text.startswith("Python"))    # True
print(text.endswith("powerful."))   # True

# Replacing substrings
print(text.replace("Python", "JavaScript"))  # JavaScript is amazing. JavaScript is powerful.
print(text.replace("Python", "Java", 1))     # Java is amazing. Python is powerful.
```

```python
Splitting and Joining
```

```python
# Split string into list
sentence = "Python is awesome"
words = sentence.split()          # Default: split by whitespace
print(words)                      # ['Python', 'is', 'awesome']

# Split with specific separator
data = "apple,banana,orange"
fruits = data.split(',')
print(fruits)                     # ['apple', 'banana', 'orange']

# Split with maxsplit
text = "one,two,three,four"
print(text.split(',', 2))         # ['one', 'two', 'three,four'] #The text.split(',', 2) method in Python splits a string into a list using a comma as the delimiter, limiting the process to a maximum of 2 splits. This results in a list with a maximum of 3 items (the split parts), where the third item contains the remainder of the string.

# Join list into string
words = ['Hello', 'World']
joined = ' '.join(words)
print(joined)                      # Hello World

# Join with different separator
print('-'.join(['2024', '02', '23']))  # 2024-02-23
```

```python
# Checking string content
print("hello123".isalnum())        # True (alphanumeric)
print("hello123".isalpha())        # False (has numbers)
print("hello".isalpha())           # True
print("123".isdigit())             # True
print("   ".isspace())             # True
print("Hello World".istitle())     # True
print("Hello123".isidentifier())   # False (can't start with alphabet)
```

```python
name = "Bob"
age = 30
score = 85.5678

# Basic f-strings
print(f"My name is {name} and I am {age} years old")

# Expressions inside f-strings
print(f"Next year I will be {age + 1}")
print(f"Score: {score:.1f}")                 # Score: 85.6
print(f"Percentage: {score/100:.2%}")        # Percentage: 85.57%

# Multi-line f-strings
message = f"""
Name: {name}
Age: {age}
Score: {score:.2f}
"""
print(message)
```

```python
def is_palindrome(text):
    """Check if a string is a palindrome (ignoring case and spaces)"""
    # Remove spaces and convert to lowercase
    cleaned = text.replace(" ", "").lower()
    return cleaned == cleaned[::-1]

# Test
print(is_palindrome("racecar"))           # True
print(is_palindrome("A man a plan a canal Panama"))  # True
print(is_palindrome("hello"))              # False #olleh
```

```python
#Anagram Checker
def are_anagrams(str1, str2):
    """Check if two strings are anagrams"""
    # Remove spaces and convert to lowercase
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    
    return sorted(str1) == sorted(str2)

# Alternative using Counter
from collections import Counter

def are_anagrams_counter(str1, str2):
    str1 = str1.replace(" ", "").lower()
    str2 = str2.replace(" ", "").lower()
    return Counter(str1) == Counter(str2)

# Test
print(are_anagrams("listen", "silent"))           # True
print(are_anagrams("hello", "world"))             # False
print(are_anagrams("Astronomer", "Moon starer"))  # True
```

```python
#Extract Email Domain
def extract_domain(email):
    """Extract domain name from email address"""
    try:
        username, domain = email.split('@')
        return f"Username: {username}, Domain: {domain}"
    except ValueError:
        return "Invalid email format"

# Test
print(extract_domain("user@example.com"))      # Username: user, Domain: example.com
print(extract_domain("invalid-email"))         # Invalid email format
print(extract_domain("AJ@gmail.com")) 
```
