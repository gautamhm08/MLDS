```python
A tuple is an ordered, immutable collection of elements in Python. Once created, you cannot modify, add, or remove elements from a tuple.

Key Characteristics:
1. Ordered: Elements maintain their position
2. Immutable: Cannot be changed after creation
3. Heterogeneous: Can contain different data types
4. Indexed: Can access elements by position
5. Hashable: Can be used as dictionary keys (if all elements are hashable)
```

```python
# Single element tuple (note the comma!)
single_tuple = (5,)  # Without comma, it's just an integer
not_a_tuple = (5)    # This is an integer, not a tuple
print(type(single_tuple))  # <class 'tuple'>
print(type(not_a_tuple))   # <class 'int'>
```

```python
# Tuples have only two built-in methods:
my_tuple = (1, 2, 3, 2, 4, 2)

# 1. count() - returns number of occurrences
print(my_tuple.count(2))  # 3

# 2. index() - returns first index of value
print(my_tuple.index(3))  # 2
# With start and end parameters
print(my_tuple.index(2, 2))  # 3 (start searching from index 2)
```

```python

What are Sets?
A set is an unordered collection of unique elements. It's mutable but can only contain immutable (hashable) elements.

Key Characteristics:

Unordered: Elements have no fixed position

Unique: No duplicate elements allowed

Mutable: Can add/remove elements

Unindexed: Cannot access by position

Fast membership testing: O(1) average time complexity
```

```python
#Creating Sets
# Empty set (note: {} creates empty dict)
empty_set = set()
print(type(empty_set))  # <class 'set'>

# Set with elements
fruits = {"apple", "banana", "cherry", "apple"}  # Duplicate will be removed
print(fruits)  # {'cherry', 'banana', 'apple'} (order may vary)

# Using set() constructor
numbers = set([1, 2, 3, 2, 4, 3])
print(numbers)  # {1, 2, 3, 4}

string_to_set = set("hello")
print(string_to_set)  # {'h', 'e', 'l', 'o'} (note: 'l' appears once)

# Set comprehension
squares = {x**2 for x in range(5)}
print(squares)  # {0, 1, 4, 9, 16}
```

```python
#Set Operations
# Adding elements
fruits = {"apple", "banana"}
fruits.add("cherry")
print(fruits)  # {'cherry', 'banana', 'apple'}

# Adding multiple elements
fruits.update(["date", "elderberry", "apple"])  # apple already exists
print(fruits)  # {'date', 'cherry', 'elderberry', 'banana', 'apple'}

# Removing elements
fruits.remove("banana")  # Raises KeyError if not found
fruits.discard("grape")  # No error if not found
removed_item = fruits.pop()  # Removes and returns arbitrary element
print(removed_item)

# Clear all elements
fruits.clear()
print(fruits)  # set()

# Membership testing (very fast!)
print("apple" in {"apple", "banana", "cherry"})  # True
print("grape" not in {"apple", "banana"})        # True
```

```python
# Define two sets for examples
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

# 1. Union (all elements from both sets)
union = set_a.union(set_b)
# or using |
union_alt = set_a | set_b
print(f"Union: {union}")  # {1, 2, 3, 4, 5, 6, 7, 8}

```

```python
# 2. Intersection (common elements)
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
intersection = set_a.intersection(set_b)
# or using &
intersection_alt = set_a & set_b
print(f"Intersection: {intersection}")  # {4, 5}

```

```python
# 3. Difference (elements in set_a but not in set_b)
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
difference = set_a.difference(set_b)
# or using -
difference_alt = set_a - set_b
print(f"Difference (A - B): {difference}")  # {1, 2, 3}

```

```python
# 4. Symmetric Difference (elements in either set, but not both)
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
sym_diff = set_a.symmetric_difference(set_b)
# or using ^
sym_diff_alt = set_a ^ set_b
print(f"Symmetric Difference: {sym_diff}")  # {1, 2, 3, 6, 7, 8}

```

```python
# 5. Subset testing
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

set_c = {1, 2, 3}
print(set_c.issubset(set_a))  # True
print(set_c <= set_a)          # True (using operator)

```

```python
# 6. Superset testing
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}
set_c = {1, 2, 3}
print(set_a.issuperset(set_c))  # True
print(set_a >= set_c)            # True

```

```python
# 7. Disjoint sets (no common elements)
set_a = {1, 2, 3, 4, 5}
set_b = {4, 5, 6, 7, 8}

set_d = {10, 11, 12}
print(set_a.isdisjoint(set_d))  # True
```

```python
#More Set Methods
# Copying sets
original = {1, 2, 3}
copied = original.copy()  # Shallow copy
copied.add(4)
print(original)  # {1, 2, 3}
print(copied)    # {1, 2, 3, 4}

# Updating sets with operations
set_x = {1, 2, 3}
set_y = {3, 4, 5}

set_x.update(set_y)  # Equivalent to union and assign
print(set_x)  # {1, 2, 3, 4, 5}

set_x = {1, 2, 3}
set_x.intersection_update(set_y)
print(set_x)  # {3}

set_x = {1, 2, 3}
set_x.difference_update(set_y)
print(set_x)  # {1, 2}

set_x = {1, 2, 3}
set_x.symmetric_difference_update(set_y)
print(set_x)  # {1, 2, 4, 5}
```
