```python
### Practice Problem 1: Extracting Email Addresses

Write a Python program that uses regular expressions to extract all valid email addresses from a given string.

**Input String:**
```

```python
import re

text = "Please contact us at support@example.com or sales@example.org for assistance. Visit our website at www.example.com."

# Regex to find email addresses
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'

found_emails = re.findall(email_pattern, text)
print(found_emails)
#In the regex pattern [A-Za-z0-9._%+-]+, the % symbol inside the square brackets [...] is treated as a literal character.

#The square brackets define a character set, and any character within that set is considered a potential match.
```

```python
### Practice Problem 2: Validating Phone Numbers

Write a Python program that uses regular expressions to validate if a given string is a valid US phone number in the format `XXX-XXX-XXXX`.

**Input Strings:**
```

```python
import re

phone_numbers = ["123-456-7890", "987.654.3210", "1234567890"]

# Regex to validate US phone numbers (XXX-XXX-XXXX)
phone_pattern = r'^\d{3}-\d{3}-\d{4}$'

for number in phone_numbers:
    if re.match(phone_pattern, number):
        print(f"{number} is a valid US phone number.")
    else:
        print(f"{number} is not a valid US phone number.")
```

```python
### Practice Problem 3: Replacing Multiple Spaces with a Single Space

Write a Python program that uses regular expressions to replace multiple consecutive spaces in a string with a single space.

**Input String:**
```

```python
import re

text = "This string  has   multiple    spaces."

# Regex to find one or more spaces and replace with a single space
cleaned_text = re.sub(r'\s+', ' ', text)
print(cleaned_text)
```

```python
1. Find all digits in a string
Text: "Order #4256 - Received on 14-03-2025 at 9:45 AM"
```

```python
import regex as re
text= "Order #4256 - Received on 14-03-2025 at 9:45 AM"
re.findall(r'\d+', text)  
# or more precisely: r'\d+' 
#['4256', '14', '03', '2025', '9', '45']
```

```python
2. Check if string starts with "Error" (case sensitive)
Text: "Error: Connection timeout"
```

```python
text= "Error: Connection timeout"
re.match(r'^Error', text) is not None
# or re.search(r'^Error', text)
```

```python
3. Extract all words that start with capital letter
Text: "Ashutosh lives in Nagpur and works at Google India"
```

```python
text= "Ramesh lives in Nagpur and works at Google India"
re.findall(r'[A-Z][a-z]+', text)  
# ['Ramesh', 'Nagpur', 'Google', 'India']
```

```python
4. Find all 10-digit mobile numbers (exactly 10 digits, no +91 or spaces)
Text: "Call me at 9876543210 or 9123456789"
```

```python
text= "Call me at 9876543210 or 9123456789"
re.findall(r'\b\d{10}\b', text)  
#['9876543210', '9123456789']
```

```python
5. Find prices in rupees (₹ followed by number)
Text: "Shirt ₹899, Jeans ₹2499, Shoes ₹1800.50"
```

```python
text = "Shirt ₹899, Jeans ₹2499, Shoes ₹1800.50"
re.findall(r'₹[\d,.]+', text)  
# ['₹899', '₹2499', '₹1800.50']
```

```python
 Remove all punctuation from a sentence (replace with space)**
Text: "Hello!!! How are you? I'm fine... thank you!!"
```

```python
text= "Hello!!! How are you? I'm fine... thank you!!"
re.sub(r'[^\w\s]', ' ', text)          # keep letters, digits, spaces
# then .strip() and re.sub(r'\s+', ' ', ...) if you want single spaces
```

```python
 Extract domain names from simple email addresses
Text: "Write to support@company.com or sales.team@shop.in"
```

```python
text= "Write to support@company.com or sales.team@shop.in"
re.findall(r'@([\w.-]+)', text)  
# ['company.com', 'shop.in']
```
