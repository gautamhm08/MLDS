```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("D:\AIT MLDS\diabetes.csv")

df
```

```python
print("First 5 rows of the DataFrame:")
print(df.head())
```

```python
print("\nDataFrame Info:")
df.info()
```

```python
print("Initial missing values:")
print(df.isnull().sum())
```

```python
import numpy as np

cols = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']

df[cols] = df[cols].replace(0, np.nan)

print("After replacing 0 with NaN:")
print(df.isnull().sum())
```

```python
for col in cols:
    df[col].fillna(df[col].median(), inplace=True)

print("Missing values after filling:")
print(df.isnull().sum())
```

```python
initial_rows = df.shape[0]

df.drop_duplicates(inplace=True)

duplicates_removed = initial_rows - df.shape[0]

print("Number of duplicate rows removed:", duplicates_removed)
```

```python
print("DataFrame Info after cleaning:")
df.info()
```

```python
print(df.describe())
```

```python
print(df['Outcome'].value_counts())
```

```python
print(df.groupby('Outcome').mean())
```

```python
sns.set_style("whitegrid")

df.hist(figsize=(12,10))
plt.show()
```

```python
sns.countplot(x='Outcome', data=df)
plt.title("Diabetes Outcome Distribution")
plt.show()
```

```python
sns.boxplot(x='Outcome', y='Glucose', data=df)
plt.title("Glucose vs Diabetes Outcome")
plt.show()
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(8,6))

sns.heatmap(correlation_matrix,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Correlation Matrix")
plt.show()
```
