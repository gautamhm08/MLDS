```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("D:\AIT MLDS\student_spending (1).csv")

df
```

```python
print(df.head())
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
initial_rows = df.shape[0]

df.drop_duplicates(inplace=True)

duplicates_removed = initial_rows - df.shape[0]

print("Duplicates removed:", duplicates_removed)
```

```python
print(df.describe())
```

```python
print("Gender Count:")
print(df['gender'].value_counts())

print("\nYear in School:")
print(df['year_in_school'].value_counts())

print("\nPayment Method:")
print(df['preferred_payment_method'].value_counts())
```

```python
#Visualizations
df[['food','housing','transportation','entertainment']].hist(figsize=(10,8))
plt.show()
```

```python
#boxplot
sns.boxplot(x='gender', y='food', data=df)

plt.title("Food Spending by Gender")
plt.show()
```

```python
#count plot
sns.countplot(x='preferred_payment_method', data=df)

plt.title("Preferred Payment Method")
plt.xticks(rotation=45)
plt.show()
```

```python
#Correlation Matrix
numeric_cols = df.select_dtypes(include=['int64','float64'])

correlation_matrix = numeric_cols.corr()

plt.figure(figsize=(10,8))

sns.heatmap(correlation_matrix,
            annot=True,
            cmap="coolwarm",
            fmt=".2f")

plt.title("Student Spending Correlation Matrix")
plt.show()
```
