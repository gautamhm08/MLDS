````python
# 4. Count vowels
word = "hello"
vowels = "aeiou"
count = 0
for char in word.lower():
    if char in vowels:
        count += 1 # count = count+1
print(f"4. Vowel count: {count}")
````

````python
# 6. Are Xs equal to Os?
string = "xoxoxx"
x_count = 0
o_count = 0
for char in string.lower():
    if char == "x":
        x_count += 1
    elif char == "o":
        o_count += 1
result = x_count == o_count
print(f"6. X equals O: {result}")
````

````python
# 10. Repeat characters
text = "now"
repeated = "" #empty string
for char in text:
    repeated += char * 2
print(f"10. Repeated: {repeated}")
````

````python
# 12. Add dots
text = "skills"
dotted = "" #empty string
for i, char in enumerate(text):
    if i < len(text) - 1:
        dotted += char + "."
    else:
        dotted += char
print(f"12. Dotted: {dotted}")
````

````python
#break — Stops the loop completely
for i in range(1, 6):
    if i == 3:
        break
    print(i)
````

````python
#continue — Skips current iteration, loop keeps going
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
````

````python
for i in range(1, 6):
    if i == 3:
        pass
    print(i)
````

````python
numbers = [1, 2, -1, 4, 5]

# break — stop if negative number found
for n in numbers:
    if n < 0:
        print("Negative found! Stopping.")
        break
    print(n)

print("---")

# continue — skip negative numbers
for n in numbers:
    if n < 0:
        continue
    print(n)

print("---")

# pass — do nothing special, print all
for n in numbers:
    if n < 0:
        pass
    print(n)
````

````python
def min_max(numbers):
    return min(numbers), max(numbers)

small, large = min_max([3, 1, 7, 2, 9]) #1,9
print("Min:", small)
print("Max:", large)
````

````python
# 1. Check even or odd
def even_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_odd(4))   # Even
print(even_odd(7))   # Odd
````

````python
# 2. Calculate area of rectangle, cirlce HW
def area(length, width):
    return length * width

print(area(5, 3))    # 15
````

````python
# 3. Find largest number , smallest HW
def largest(a, b, c):
    if a > b and a > c:
        return a
    elif b > c:
        return b
    else:
        return c

print(largest(10, 25, 15))   # 25
````

````python
Common Mistakes
# ❌ Forgetting to call the function
def greet():
    print("Hello")
# Nothing happens if you don't call it!

# ✅ Correct
greet()

# ❌ Forgetting return
def add(a, b):
    a + b          # calculated but not returned

result = add(3, 4)
print(result)      # None

# ✅ Correct
def add(a, b):
    return a + b
````
