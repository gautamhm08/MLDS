```python
#Identity Operators
#In Python, is and is not are the identity operators both are used to check if two values are located on the same part of the memory. Two variables that are equal do not imply that they are identical. 

#is          True if the operands are identical 
#is not      True if the operands are not identical 
```

```python
a = 10
b = 20
c = a

print(a is not b)
print(a is c)
print(a is b)
```

```python
#Membership Operators
#In Python, in and not in are the membership operators that are used to test whether a value or variable is in a sequence.

#in            True if value is found in the sequence
#not in        True if value is not found in the sequence
```

```python
x = 24
y = 20
list = [10, 20, 30, 40, 50]

if (x not in list):
    print("x is NOT present in given list")
else:
    print("x is present in given list")

if (y in list):
    print("y is present in given list")
else:
    print("y is NOT present in given list")
```

```python
"""
Simple Calculator - Python Version
Shows all calculator operations using Python operators
"""

def display_menu():
    """Display calculator menu"""
    print("=" * 40)
    print("SIMPLE CALCULATOR - PYTHON VERSION")
    print("=" * 40)
    print("1. Addition (+)")
    print("2. Subtraction (-)")
    print("3. Multiplication (*)")
    print("4. Division (/)")
    print("5. Floor Division (//)")
    print("6. Modulus (%)")
    print("7. Power (**)")
    print("=" * 40)

def calculate(num1, num2, choice):
    """Perform calculation based on user choice"""
    if choice == 1:
        result = num1 + num2
        operation = "+"
    elif choice == 2:
        result = num1 - num2
        operation = "-"
    elif choice == 3:
        result = num1 * num2
        operation = "*"
    elif choice == 4:
        if num2 != 0:
            result = num1 / num2
            operation = "/"
        else:
            return None, "Error: Division by zero!"
    elif choice == 5:
        if num2 != 0:
            result = num1 // num2
            operation = "//"
        else:
            return None, "Error: Division by zero!"
    elif choice == 6:
        if num2 != 0:
            result = num1 % num2
            operation = "%"
        else:
            return None, "Error: Division by zero!"
    elif choice == 7:
        result = num1 ** num2
        operation = "**"
    else:
        return None, "Invalid choice!"
    
    return result, f"{num1} {operation} {num2} = {result}"

def demo_calculations():
    """Show demo calculations with different operators"""
    print("\n" + "=" * 40)
    print("DEMO CALCULATIONS")
    print("=" * 40)
    
    # Using different numbers for demo
    a, b = 15, 4
    
    print(f"Numbers: {a} and {b}")
    print(f"Addition: {a} + {b} = {a + b}")
    print(f"Subtraction: {a} - {b} = {a - b}")
    print(f"Multiplication: {a} * {b} = {a * b}")
    print(f"Division: {a} / {b} = {a / b}")
    print(f"Floor Division: {a} // {b} = {a // b}")
    print(f"Modulus: {a} % {b} = {a % b}")
    print(f"Power: {a} ** {b} = {a ** b}")
    
    # Comparison operators
    print("\n" + "=" * 40)
    print("COMPARISON OPERATORS")
    print("=" * 40)
    print(f"{a} > {b} ? {a > b}")
    print(f"{a} < {b} ? {a < b}")
    print(f"{a} == {b} ? {a == b}")
    print(f"{a} != {b} ? {a != b}")
    print(f"{a} >= {b} ? {a >= b}")
    print(f"{a} <= {b} ? {a <= b}")

def interactive_mode():
    """Interactive calculator mode"""
    while True:
        display_menu()
        
        try:
            choice = int(input("\nEnter your choice (1-7) or 0 to exit: "))
            
            if choice == 0:
                print("Thank you for using Python Calculator!")
                break
            
            if choice < 1 or choice > 7:
                print("Please enter valid choice (1-7)!")
                continue
            
            num1 = float(input("Enter first number: "))
            num2 = float(input("Enter second number: "))
            
            result, message = calculate(num1, num2, choice)
            
            if result is None:
                print(f"\n❌ {message}")
            else:
                print(f"\n✅ {message}")
                
                # Show additional info
                print(f"\nResult type: {type(result)}")
                if isinstance(result, float):
                    print(f"Rounded: {round(result, 2)}")
                
        except ValueError:
            print("❌ Please enter valid numbers!")
        except KeyboardInterrupt:
            print("\n\nGoodbye!")
            break

# Main program
if __name__ == "__main__":
    print("\n🐍 PYTHON CALCULATOR PROGRAM")
    print("This demonstrates all operators you've learned!\n")
    
    # Show demo calculations
    demo_calculations()
    
    # Ask if user wants interactive mode
    print("\n" + "=" * 40)
    response = input("Do you want to try interactive mode? (yes/no): ").lower()
    
    if response == 'yes' or response == 'y':
        interactive_mode()
    else:
        print("\nYou can see the calculator web version now!")
        print("Open index.html in your browser!")
```
