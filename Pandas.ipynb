```python
#Selection & Indexing
#Problem 9: loc vs iloc
import pandas as pd

df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': ['a', 'b', 'c', 'd', 'e'],
    'C': [10, 20, 30, 40, 50]
}, index=['row1', 'row2', 'row3', 'row4', 'row5'])

print("Original DataFrame:")
print(df)
print()

# 1. Row 'row3' using loc
print("1. df.loc['row3']:")
print(df.loc['row3'])
print()

# 2. Row at position 2 using iloc
print("2. df.iloc[2]:")
print(df.iloc[2])
print()

# 3. Rows 'row2' through 'row4', columns 'A' and 'C' using loc
print("3. df.loc['row2':'row4', ['A', 'C']]:")
print(df.loc['row2':'row4', ['A', 'C']])
print()

# 4. Rows at positions 1-3, columns at positions 0-1 using iloc
# Note: iloc end is EXCLUSIVE (positions 1, 2, 3 not included for rows)
print("4. df.iloc[1:4, 0:2]:")
print(df.iloc[1:4, 0:2])

# Output:
# 3. df.loc['row2':'row4', ['A', 'C']]:
#        A   C
# row2   2  20
# row3   3  30
# row4   4  40
#
# 4. df.iloc[1:4, 0:2]:
#        A  B
# row2   2  b
# row3   3  c
# row4   4  d
```

```python
#Statistics
# Descriptive Statistics
import pandas as pd
import numpy as np

# Sample data
df = pd.DataFrame({
    'Salary': [50000, 60000, 55000, 80000, 52000, 58000, 120000, 51000],
    'Years_Experience': [1, 3, 2, 5, 2, 3, 10, 1]
})

# 1. Calculate mean, median, std for both columns
stats = df.agg(['mean', 'median', 'std'])
print("Basic statistics:")
print(stats)
print()

# 2. Find correlation between salary and experience
correlation = df['Salary'].corr(df['Years_Experience'])
print(f"Correlation: {correlation:.3f}")
print()

# 3. Identify outliers using IQR method
def find_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]

print("Outliers in Salary:")
print(find_outliers_iqr(df['Salary']))
```

```python
#Missing Data
# Handling Missing Values
import pandas as pd
import numpy as np

# Sample data with missing ages
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David', 'Eve', 'Frank'],
    'Age': [15, 16, np.nan, 17, np.nan, 15],
    'Grade_Level': [9, 10, 11, 12, 9, 9]
})

print("Original with missing values:")
print(df)
print(f"\nMissing values: {df['Age'].isnull().sum()}")

# 1. Fill missing ages with median age grouped by Grade_Level
df['Age'] = df.groupby('Grade_Level')['Age'].transform(
    lambda x: x.fillna(x.median())
)

print("\nAfter filling with median by Grade_Level:")
print(df)

# 2. Verify no missing values remain
print(f"\nMissing values after fill: {df['Age'].isnull().sum()}")

# 3. Why median? It's robust to outliers. If one student was 50 years old,
#    the mean would be skewed, but median represents typical student age.
```

```python
#Filtering & Sorting
#Complex Filtering
import pandas as pd

# Sample sales data
df = pd.DataFrame({
    'Revenue': [1500, 800, 3000, 6000, 2500, 4500, 900],
    'Region': ['North', 'South', 'East', 'North', 'South', 'West', 'North'],
    'Product_Category': ['Electronics', 'Clothing', 'Food', 'Clothing',
                         'Electronics', 'Food', 'Clothing']
})

print("Original data:")
print(df)
print()

# Filter: Revenue 1000-5000 AND Region in (North, South) AND NOT Electronics
filtered = df[
    (df['Revenue'].between(1000, 5000)) &
    (df['Region'].isin(['North', 'South'])) &
    (df['Product_Category'] != 'Electronics')
]

print("Filtered results:")
print(filtered)
```

```python
#Sorting
import pandas as pd

# Sample employee data
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Carol', 'David'],
    'Department': ['IT', 'HR', 'IT', 'Finance'],
    'Salary': [70000, 50000, 80000, 60000]
})

print("Original:")
print(df)
print()

# Sort by Department (asc), then Salary (desc)
df_sorted = df.sort_values(
    by=['Department', 'Salary'],
    ascending=[True, False]
)

print("Sorted:")
print(df_sorted)
print()

# Reset index
df_sorted = df_sorted.reset_index(drop=True)
print("With reset index:")
print(df_sorted)
```

```python
#GroupBy Aggregation
import pandas as pd

# Employee data
df = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
    'Role': ['Dev', 'Manager', 'Recruiter', 'Manager', 'Analyst', 'Manager'],
    'Salary': [80000, 120000, 60000, 90000, 70000, 110000],
    'Years_at_Company': [2, 5, 1, 4, 3, 6]
})

print("Employee data:")
print(df)
print()

# 1. Average salary by department
avg_salary = df.groupby('Department')['Salary'].mean()
print("1. Average Salary by Department:")
print(avg_salary)
print()

# 2. Max years at company by role within each department
max_years = df.groupby(['Department', 'Role'])['Years_at_Company'].max()
print("2. Max Years by Department and Role:")
print(max_years)
print()

# 3. Count employees per department
emp_count = df.groupby('Department').size()
print("3. Employee Count by Department:")
print(emp_count)
```

```python
#Multiple Aggregations
import pandas as pd

df = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance'],
    'Salary': [80000, 120000, 60000, 90000, 70000, 110000],
    'Years_at_Company': [2, 5, 1, 4, 3, 6]
})

# Multiple aggregations with .agg()
result = df.groupby('Department').agg({
    'Salary': ['sum', 'mean', 'count'],
    'Years_at_Company': 'max'
})

# Flatten column names
result.columns = ['Total_Salary', 'Avg_Salary', 'Emp_Count', 'Max_Years']
result = result.reset_index()

print(result)

# Output:
#   Department  Total_Salary  Avg_Salary  Emp_Count  Max_Years
# 0    Finance        180000     90000.0          2          6
# 1         HR        150000     75000.0          2          4
# 2         IT        200000    100000.0          2          5
```
