```python
Lambda functions, also known as anonymous functions, are a way to create small, one-line functions without using the def keyword. They're useful for short, throwaway functions that you don't need to reuse extensively. Lambda functions are often used in situations where a function is required as an argument to another function, like in sorting, mapping, or filtering data.

Syntax

The basic syntax of a lambda function is:

lambda arguments: expression

arguments: Zero or more comma-separated parameters (e.g., x, x, y).
expression: A single expression that gets evaluated and returned (no statements like if blocks or loops allowed directly; use ternary if needed).

Unlike regular functions defined with def, lambdas don't have a name (hence "anonymous") and are limited to a single expression.
```

```python
#No arguments
no_arg = lambda: "Hello, World!"
print(no_arg())  # Output: Hello, World!
```

```python
#Single argument:
square = lambda x: x ** 2
print(square(5))  # Output: 25
```

```python
#Multiple arguments
add = lambda x, y: x + y
print(add(3, 4))  # Output: 7
```

```python
#With conditional (ternary operator)
max_num = lambda a, b: a if a > b else b
print(max_num(10, 20))  # Output: 20
```

```python
Using Lambda with map()
map() applies a function to each item in an iterable (like a list) and returns a map object (which you can convert to a list).
```

```python
# Square each number
numbers = [1,2,3,4,5,6]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16, 25, 36]
```

```python
#Map with multiple iterables:
evens = [2, 4, 6, 8]
odds = [1, 3, 5, 7]
# Add corresponding elements
sums = list(map(lambda x, y: x + y, evens, odds))
print(sums)  # Output: [3, 7, 11, 15]
```

```python
#Map with strings
words = ["apple", "banana", "cherry"]
uppercased = list(map(lambda s: s.upper(), words))
print(uppercased)  # Output: ['APPLE', 'BANANA', 'CHERRY']
```

```python
Using Lambda with filter()

filter() applies a function that returns True or False to each item and keeps only those where the function returns True.
```

```python
# Keep even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)  # Output: [2, 4, 6]
```

```python
#Filter with conditionals:
# Numbers greater than 3
greater_than_three = list(filter(lambda x: x > 3, numbers))
print(greater_than_three)  # Output: [4, 5, 6]
```

```python
#Filter non-numeric data
fruits = ["apple", "banana", "cherry", "date"]
# Words starting with 'b' or 'c'
bc_fruits = list(filter(lambda f: f[0] in ['b', 'c'], fruits))
print(bc_fruits)  # Output: ['banana', 'cherry']
```

```python
#Using Lambda with sorted()

#sorted() sorts an iterable, and you can provide a key function (often a lambda) to customize the sort.

#Basic sort with lambda key:
numbers = [5,6,1,4,2,3]
# Sort by value (default is ascending)
sorted_numbers = sorted(numbers, key=lambda x: x)  # Same as sorted(numbers)
print(sorted_numbers)  # Output: [1, 2, 3, 4, 5, 6]
```

```python
#Side-by-side examples (most common patterns)
```

```python
# 1. Square numbers > 10
numbers = [8, 12, 3, 25, 7, 19]

# List comprehension ──────── preferred
squared = [x*x for x in numbers if x > 10]

# map + filter ────────────── more verbose
squared = list(map(lambda x: x*x, filter(lambda x: x > 10, numbers)))
```

```python
# 2. Get even squares
# Comprehension (clean)
even_squares = [x*x for x in numbers if x % 2 == 0 and x*x > 50]

# Lambda way (painful)
even_squares = list(
    map(lambda x: x*x,
        filter(lambda x: x % 2 == 0 and x*x > 50, numbers))
)
```

```python
# 3. Uppercase names longer than 4 chars
names = ["ana", "Emma", "Liam", "Noah", "Olivia"]

# Comprehension
long_upper = [name.upper() for name in names if len(name) > 4]

# map + filter
long_upper = list(map(str.upper, filter(lambda n: len(n) > 4, names)))
```
