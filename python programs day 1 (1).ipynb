```python
#WAP To swap two variables Using a Third Variable
# Initialize two variables
a = 10
b = 20
print("Before swapping: a =", a, "b =", b)
# Using a third variable (temp) to swap
temp = a
a = b
b = temp
print("After swapping: a =", a, "b =", b)   
```

```python
#Without Using a Third Variable
# Initialize two variables
a = 10
b = 20
print("Before swapping: a =", a, "b =", b)
# Swapping using tuple unpacking (no third variable needed)
a, b = b, a
print("After swapping: a =", a, "b =", b) 
```

```python
#Without Using a Third Variable
# Initialize two variables
a = 10
b = 20
print("Before swapping: a =", a, "b =", b)
# Swapping using Addition/Subtraction (no third variable needed)
a = a + b
b = a - b
a = a - b
print("After swapping: a =", a, "b =", b)   
```

```python
#Find Largest of Two Numbers

# Method 1: Using if-else
a, b = 10, 20
if a > b:
    largest = a
else:
    largest = b
print(f"Largest: {largest}")

# Method 2: Using max() function
a, b = 10, 20
largest = max(a, b)
print(f"Largest: {largest}"
```

```python
# Try on your own Find Minimum of Three Numbers

# Method 1: Using if-else ladder
a, b, c = 15, 7, 22
if a <= b and a <= c:
    minimum = a
elif b <= a and b <= c:
    minimum = b
else:
    minimum = c
print(f"Minimum: {minimum}")

# Method 2: Using min() function
a, b, c = 15, 7, 22
minimum = min(a, b, c)
print(f"Minimum: {minimum}")
```

```python
#Factorial Method 1: Using loop
n = 5
factorial = 1
for i in range(1, n+1):
    factorial *= i # factorial=factorial*i
print(f"Factorial of {n}: {factorial}")

# Method 2: Using recursion
def factorial_recursive(n):
    if n <= 1:
        return 1
    return n * factorial_recursive(n-1)

n = 5
print(f"Factorial of {n}: {factorial_recursive(n)}")
```

```python
#Check Prime Number

num = 17
is_prime = True
if num < 2:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
print(f"{num} is prime: {is_prime}")
```

```python
# Pattern 1: Right Triangle with 1's
print("1. Triangle with all 1's:")
for i in range(1, 6):
    for j in range(i):
        print("1", end=" ")
    print()
print()
```

```python
# Pattern 2: Right Triangle with increasing numbers
print("2. Triangle with increasing numbers (1 to row number):")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
print()
```

```python
# Pattern 3: Right Triangle with same numbers in each row
print("3. Triangle with same numbers in each row:")
for i in range(1, 6):
    for j in range(i):
        print(i, end=" ")
    print()
print()
```

```python
# Pattern 4: Inverted Right Triangle (reverse order)
print("4. Inverted triangle (5 to 1):")
for i in range(5, 0, -1):
    for j in range(i):
        print(j + 1, end=" ")
    print()
print()
```

```python
# Pattern 5: Pascal's Triangle style (without actual Pascal calculation)
print("5. Triangle with repeating row numbers:")
for i in range(1, 6):
    for j in range(i):
        print(i if j % 2 == 0 else j + 1, end=" ")
    print()
print()
```

```python
# Pattern 6: Triangle with alternating 1 and 0
print("6. Triangle with alternating 1 and 0:")
for i in range(1, 6):
    for j in range(i):
        print("1" if j % 2 == 0 else "0", end=" ")
    print()
print()
```

```python
# Pattern 7: Triangle with binary pattern
print("7. Triangle with binary pattern based on position:")
for i in range(1, 6):
    for j in range(i):
        print((i + j) % 2, end=" ")
    print()
print()
```

```python
# Pattern 8: Pyramid pattern (centered numbers)
print("8. Pyramid pattern with numbers:")
for i in range(1, 6):
    # Print spaces
    for j in range(5 - i):
        print(" ", end=" ")
    # Print numbers
    for k in range(1, i + 1):
        print(k, end=" ")
    print()
print()
```

```python
# Pattern 9: Triangle with even numbers
print("9. Triangle with even numbers:")
for i in range(1, 6):
    for j in range(i):
        print((j + 1) * 2, end=" ")
    print()
print()
```

```python
# Pattern 10: Triangle with odd numbers
print("10. Triangle with odd numbers:")
for i in range(1, 6):
    for j in range(i):
        print((j * 2) + 1, end=" ")
    print()
print()
```

```python
# Bonus Pattern 11: Floyd's Triangle
print("Bonus: Floyd's Triangle:")
num = 1
for i in range(1, 6):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```
