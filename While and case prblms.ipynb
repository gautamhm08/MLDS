```python
# Basic while loop
count = 1
while count <= 5:
    print(f"Count: {count}")
    count += 1
# Output: Count: 1,2,3,4,5
```

```python
#While vs For Loop - Same Output, Different Approach
# FOR LOOP (already taught)
print("For loop:")
for i in range(1, 6):
    print(i, end=" ")

print("\n")

# WHILE LOOP (new)
print("While loop:")
i = 1
while i <= 5:
    print(i, end=" ")
    i += 1
```

```python
#Key Differences - When to Use Which
```

```python
# FOR LOOP - Best when you know the number of iterations
print("Printing 10 numbers - known iterations:")
for i in range(10):
    print(i, end=" ")

print("\n")

# WHILE LOOP - Best when iterations depend on a condition
print("Keep doubling until > 100 - unknown iterations:")
num = 2
while num <= 100:
    print(num, end=" ")
    num *= 2
# Output: 2 4 8 16 32 64
```

```python

# Keep adding digits until we get single digit
num = 9875
print(f"Starting number: {num}")
while num >= 10:
    total = 0
    temp = num
    while temp > 0:
        total += temp % 10 # total = total + temp%10
        temp //= 10 # this is floor division i.e. 5 // 2 gives Result=2	 How? -> 5 divided by 2 is 2.5, rounded down is 2.
    num = total
    print(f"Sum of digits: {num}")
print(f"Final single digit: {num}")
```

```python
#Menu-driven program
# Perfect example for while loop - runs until user chooses to exit
choice = 0
while choice != 5:
    print("\n--- MENU ---")
    print("1. Say Hello")
    print("2. Print numbers 1-5")
    print("3. Print current time")
    print("4. Tell a joke")
    print("5. Exit")
    
    choice = int(input("Enter your choice: "))
    
    if choice == 1:
        print("Hello! How are you?")
    elif choice == 2:
        for i in range(1, 6):
            print(i, end=" ")
    elif choice == 3:
        import datetime # import library called datetime
        print("Current time:", datetime.datetime.now().strftime("%H:%M:%S"))
    elif choice == 4:
        print("Why don't scientists trust atoms? Because they make up everything!")
    elif choice == 5:
        print("Goodbye!")
    else:
        print("Invalid choice!")
```

```python
# Keep asking until correct password
password = "python123"
attempt = ""
attempt_count = 0

while attempt != password and attempt_count < 3:
    attempt = input("Enter password: ")
    attempt_count += 1
    if attempt == password:
        print("Access granted!")
    elif attempt_count < 3:
        print(f"Wrong password. Attempts left: {3 - attempt_count}")
    else:
        print("Too many attempts. Access locked!")
```

```python
#Break and Continue in While Loops
```

```python
# Break - exit loop immediately
print("Break example - stop at 5:")
num = 1
while num <= 10:
    if num == 5:
        print("Found 5! Breaking...")
        break
    print(num)
    num += 1

print("\n")

# Continue - skip current iteration
print("Continue example - skip 5:")
num = 0
while num < 10:
    num += 1
    if num == 5:
        print("Skipping 5...")
        continue
    print(num)
```

```python
#Infinite Loops (and how to avoid them)
```

```python
# Common mistake - forgetting to update counter
# DON'T DO THIS:
# i = 1
# while i <= 5:
#     print(i)
#     # forgot i += 1 - will run forever!

# Correct way - always update the condition variable
i = 1
while i <= 5:
    print(i)
    i += 1  # Don't forget this!
```

```python
#Number Guessing Game
import random # this is a library from which we can import the things we require
secret = random.randint(1, 20)
guess = 0
attempts = 0

print("I'm thinking of a number between 1 and 20")
while guess != secret:
    guess = int(input("Your guess: "))
    attempts += 1
    
    if guess < secret:
        print("Too low! Try again.")
    elif guess > secret:
        print("Too high! Try again.")
    else:
        print(f"Correct! You got it in {attempts} attempts!")
```

```python
#Simple Calculator with while loop
```

```python
# Calculator that keeps running until user quits
while True: # example of infinite loop
    print("\n--- Simple Calculator ---")
    print("1. Add")
    print("2. Subtract")
    print("3. Multiply")
    print("4. Divide")
    print("5. Exit")
    
    choice = input("Choice (1-5): ")
    
    if choice == '5':
        print("Goodbye!")
        break
    
    if choice in ['1', '2', '3', '4']:
        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))
        
        if choice == '1':
            print(f"Result: {num1 + num2}")
        elif choice == '2':
            print(f"Result: {num1 - num2}")
        elif choice == '3':
            print(f"Result: {num1 * num2}")
        elif choice == '4':
            if num2 != 0:
                print(f"Result: {num1 / num2}")
            else:
                print("Error: Cannot divide by zero!")
    else:
        print("Invalid choice!")
```

```python
#Reverse a Number
```

```python
num = int(input("Enter a number: "))
original = num
reversed_num = 0

while num > 0:
    digit = num % 10 # 234%10 = 4 # 3
    reversed_num = reversed_num * 10 + digit # 0 = 0 +4 # 40+3=43
    num //= 10 #23 #2

print(f"Reverse of {original} is {reversed_num}")
```

```python
#Fibonacci Series using while
```

```python
n = int(input("How many Fibonacci numbers? "))
a, b = 0, 1
count = 0

print("Fibonacci Series:")
while count < n:
    print(a, end=" ")
    a, b = b, a + b
    count += 1
```
