```python
1. Introduction to Data Cleaning
First, let's import pandas and create a sample dataset with common issues:
```

```python
import pandas as pd
import numpy as np

# Create a sample dataset with various data quality issues
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', np.nan, 'Grace'],
    'Age': [25, np.nan, 30, 35, 29, 40, np.nan],
    'Salary': [50000, 60000, np.nan, 80000, 55000, 75000, 65000],
    'Department': ['HR', 'IT', 'IT', np.nan, 'HR', 'Finance', 'IT'],
    'Join_Date': ['2020-01-15', '2019-03-20', '2021-06-01', 
                  '2018-11-30', '2022-02-10', '2020-07-15', '2019-09-01']
}

df = pd.DataFrame(data)
print("Original Dataset:")
print(df)
print("\nDataset Info:")
print(df.info())
```

```python
2. Identifying Missing Data
Let's learn how to detect missing values:
```

```python
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
# Visual representation of missing data
print("\nBoolean mask of missing values:")
print(df.isnull())
```

```python
# Check percentage of missing data
print("\nPercentage of missing data:")
print((df.isnull().sum() / len(df)) * 100)
```

```python
# Check which rows have any missing values
print("\nRows with any missing values:")
print(df[df.isnull().any(axis=1)])
```

```python
# Check which rows have all values present
print("\nComplete rows (no missing values):")
print(df[df.notnull().all(axis=1)])
```

```python
3. Handling Missing Data
Method 1: Dropping Missing Data
```

```python
# Drop rows with any missing values
print(df)
df_dropped_rows = df.dropna()
print("***********")
print("After dropping rows with any missing values:")
print(df_dropped_rows)
print(f"Rows dropped: {len(df) - len(df_dropped_rows)}")
```

```python

# Drop columns with missing values
df_dropped_cols = df.dropna(axis=1)
print("\nAfter dropping columns with any missing values:")
print(df_dropped_cols)

```

```python
# Drop rows where specific column has missing values
df_dropped_specific = df.dropna(subset=['Age', 'Salary'])
print("\nAfter dropping rows missing Age or Salary:")
print(df_dropped_specific)

```

```python
# Drop rows with threshold (keep rows with at least 3 non-null values)
df_threshold = df.dropna(thresh=3)
print("\nKeep rows with at least 3 non-null values:")
print(df_threshold)
```

```python
Method 2: Filling Missing Data
```

```python
# Create a fresh copy for filling examples
df_filled = df.copy()

# Fill all missing values with a specific value
df_filled_fillna = df_filled.fillna('Unknown')
print("Fill all missing with 'Unknown':")
print(df_filled_fillna)

```

```python
When to use mean, mode and median ?
```

```python
# Fill with different values for different columns
df_filled = df.copy()
df_filled['Age'] = df_filled['Age'].fillna(df_filled['Age'].mean())
df_filled['Salary'] = df_filled['Salary'].fillna(df_filled['Salary'].median())
df_filled['Department'] = df_filled['Department'].fillna('Unknown')
df_filled['Name'] = df_filled['Name'].fillna('Anonymous')

print("\nAfter filling with column-specific values:")
print(df_filled)

```

```python
# Forward fill (carry forward previous value)
df_ffill = df.copy()
df_ffill = df_ffill.fillna(method='ffill')
print("\nForward fill:")
print(df_ffill)

```

```python
# Backward fill (use next value)
df_bfill = df.copy()
df_bfill = df_bfill.fillna(method='bfill')
print("\nBackward fill:")
print(df_bfill)
```

```python
# Interpolation (for numeric data)
df_interpolate = df.copy()
df_interpolate['Age'] = df_interpolate['Age'].interpolate()
df_interpolate['Salary'] = df_interpolate['Salary'].interpolate()
print("\nAfter interpolation:")
print(df_interpolate)
```

```python
Handling Duplicates and Data Types
```

```python
# Create dataset with duplicates
data_with_dupes = {
    'ID': [1, 2, 3, 4, 2, 5, 1],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Bob', 'Eve', 'Alice'],
    'Score': [85, 92, 78, 90, 92, 88, 85]
}
df_dupes = pd.DataFrame(data_with_dupes)

print("Dataset with duplicates:")
print(df_dupes)

# Find duplicates
print("\nDuplicate rows:")
print(df_dupes[df_dupes.duplicated()])
```

```python
# Find duplicates based on specific columns
print("\nDuplicate based on ID only:")
print(df_dupes[df_dupes.duplicated(subset=['ID'], keep='first')])
```

```python
# Drop duplicates
df_no_dupes = df_dupes.drop_duplicates()
print("\nAfter removing duplicates:")
print(df_no_dupes)
```

```python
# Keep last occurrence
df_no_dupes_last = df_dupes.drop_duplicates(keep='last')
print("\nKeep last occurrence:")
print(df_no_dupes_last)
```

```python
Data Type Conversion and String Cleaning
```

```python
# Create dataset with mixed data types
data_mixed = {
    'Date': ['2023-01-01', '2023-02-01', 'invalid_date', '2023-04-01'],
    'Price': ['$100', '$200', '$150', 'invalid'],
    'Quantity': ['10', '20', '30', '40'],
    'Name': ['  Alice  ', 'BOB', 'charlie', '  DAVID  ']
}
df_mixed = pd.DataFrame(data_mixed)

print("Dataset with mixed types:")
print(df_mixed)
```

```python
# Convert date column
df_mixed['Date'] = pd.to_datetime(df_mixed['Date'], errors='coerce')
print("\nAfter date conversion (invalid becomes NaT):")
print(df_mixed)
```

```python
# Clean price column
df_mixed['Price'] = df_mixed['Price'].str.replace('$', '')
df_mixed['Price'] = pd.to_numeric(df_mixed['Price'], errors='coerce')
print("\nAfter price cleaning:")
print(df_mixed)
```

```python
# Clean string columns
df_mixed['Name'] = df_mixed['Name'].str.strip()
df_mixed['Name'] = df_mixed['Name'].str.title()
print("\nAfter string cleaning:")
print(df_mixed)
```

```python
# Convert quantity to numeric
df_mixed['Quantity'] = pd.to_numeric(df_mixed['Quantity'])
print("\nFinal cleaned dataset:")
print(df_mixed)
print(df_mixed.info())
```

```python
# Create a realistic employee dataset
np.random.seed(42)
n_rows = 100

employee_data = {
    'Employee_ID': range(1001, 1001 + n_rows),
    'Name': [f'Employee_{i}' for i in range(n_rows)],
    'Age': np.random.randint(22, 65, n_rows),
    'Salary': np.random.randint(40000, 120000, n_rows),
    'Years_Experience': np.random.randint(0, 40, n_rows),
    'Department': np.random.choice(['IT', 'HR', 'Finance', 'Marketing', 'Sales'], n_rows),
    'Performance_Score': np.random.uniform(1, 5, n_rows).round(2)
}

# Intentionally introduce missing values
df_employee = pd.DataFrame(employee_data)

# Add some missing values
missing_indices = np.random.choice(df_employee.index, size=15, replace=False)
df_employee.loc[missing_indices[:5], 'Age'] = np.nan
df_employee.loc[missing_indices[5:10], 'Salary'] = np.nan
df_employee.loc[missing_indices[10:], 'Performance_Score'] = np.nan

print("Employee Dataset Overview:")
print(df_employee.head())
print("\nDataset Info:")
print(df_employee.info())

# Comprehensive cleaning pipeline
def clean_employee_data(df):
    """
    Complete data cleaning pipeline for employee data
    """
    df_clean = df.copy()
    
    # 1. Check missing data
    print("Missing data before cleaning:")
    print(df_clean.isnull().sum())
    
    # 2. Handle missing values
    # Fill Age with median (less sensitive to outliers)
    df_clean['Age'] = df_clean['Age'].fillna(df_clean['Age'].median())
    
    # Fill Salary with median by department
    df_clean['Salary'] = df_clean.groupby('Department')['Salary'].transform(
        lambda x: x.fillna(x.median())
    )
    
    # Fill Performance Score with mean
    df_clean['Performance_Score'] = df_clean['Performance_Score'].fillna(
        df_clean['Performance_Score'].mean()
    )
    
    # 3. Remove any duplicates
    df_clean = df_clean.drop_duplicates(subset=['Employee_ID'])
    
    # 4. Validate data types
    df_clean['Employee_ID'] = df_clean['Employee_ID'].astype(int)
    df_clean['Age'] = df_clean['Age'].astype(int)
    df_clean['Salary'] = df_clean['Salary'].astype(int)
    
    # 5. Check for outliers (optional)
    # Remove salaries that are unrealistic (e.g., below 20000 or above 500000)
    df_clean = df_clean[(df_clean['Salary'] >= 20000) & (df_clean['Salary'] <= 500000)]
    
    print("\nMissing data after cleaning:")
    print(df_clean.isnull().sum())
    
    return df_clean

# Apply cleaning
df_employee_clean = clean_employee_data(df_employee)

print("\nCleaned Dataset Info:")
print(df_employee_clean.info())
print("\nFirst few rows of cleaned data:")
print(df_employee_clean.head())
```
