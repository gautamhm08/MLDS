```python
# Add / update
person["name"] = "Neil"
person["age"] = 17
person["hobbies"] = ["cricket", "coding", "memes"]
print(person)
```

```python
# Access value
print(person["name"])          # Neil
```

```python
# Get with safe access (very important!)
print(person.get("height"))          # None (no error!)
print(person.get("height", 160))     # 160 (default value)
```

```python
#Access the entire hobbies list
# Method 1: Using square brackets
all_hobbies = person["hobbies"]
print(all_hobbies)           # ['cricket', 'coding', 'memes']
print(type(all_hobbies))     # <class 'list'>

# Method 2: Using get() (safer)
all_hobbies = person.get("hobbies")
print(all_hobbies)           # ['cricket', 'coding', 'memes']
```

```python
# Remove
person.pop("grade")             # removes & returns value
del person["hobbies"]           # just removes
```

```python
#accessing dictionary keys, values and items
my_dict = {"a": 1, "b": 2, "c": 3}

# Get all keys
print(my_dict.keys())     # dict_keys(['a', 'b', 'c'])

# Get all values
print(my_dict.values())   # dict_values([1, 2, 3])

# Get all key-value pairs
print(my_dict.items())    # dict_items([('a', 1), ('b', 2), ('c', 3)])

# Check if key exists
print("a" in my_dict)     # True
print("z" in my_dict)     # False

# Get length
print(len(my_dict))       # 3
```

```python
scores = {"Rahul": 87, "Sneha": 95, "Vikram": 64, "Priya": 92}

# 1. Loop over keys (most common)
for name in scores:
    print(name, "→", scores[name])
```

```python
# 2. Loop over keys explicitly
for name in scores.keys():
    print(name)
```

```python
# 3. Loop over values
for mark in scores.values():
    print(mark, end=" ")
```

```python
# 4. Loop over both key and value (very powerful!)
for name, mark in scores.items():
    print(f"{name:8} : {mark}")
    if mark >= 90:
        print("   → Outstanding!")
```

```python
student = {"name": "Alice", "age": 20, "grade": "A"}

# Iterating through keys
print("Keys:")
for key in student:
    print(f"  {key}")

# Iterating through values
print("\nValues:")
for value in student.values():
    print(f"  {value}")

# Iterating through key-value pairs
print("\nKey-Value pairs:")
for key, value in student.items():
    print(f"  {key}: {value}")
```

```python
#Nested Dictionaries
# Dictionary containing dictionaries
university = {
    "student1": {
        "name": "Alice",
        "age": 20,
        "courses": ["Math", "Physics"]
    },
    "student2": {
        "name": "Bob",
        "age": 22,
        "courses": ["Chemistry", "Biology"]
    }
}

# Accessing nested values
print(university["student1"]["name"])        # Alice
print(university["student2"]["courses"][0])  # Chemistry

# Adding to nested structure
university["student1"]["grade"] = "A"
print(university["student1"])
```

```python
#Practical Examples
#Example 1: Word Counter
def word_counter(text):
    """Count frequency of each word in text"""
    words = text.lower().split()
    word_count = {}

    for word in words:
        word_count[word] = word_count.get(word, 0) + 1

    return word_count

sentence = "the cat and the dog and the bird"
result = word_counter(sentence)
print(result)  # {'the': 3, 'cat': 1, 'and': 2, 'dog': 1, 'bird': 1}
```

```python
#Problem 3: Word Lengths

words = ["cat", "elephant", "dog"]

word_lengths = {}

for word in words:
    word_lengths[word] = len(word)

print(word_lengths)
```

```python
#Problem 4: Character Counter Function

def count_characters(text):
    result = {}
    
    for char in text:
        if char != " ":   # ignore spaces
            if char in result:
                result[char] += 1
            else:
                result[char] = 1
                
    return result


print(count_characters("hello world"))
```
